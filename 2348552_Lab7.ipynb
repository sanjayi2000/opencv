```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r"C:\Users\sanja\Downloads\GP2403_045106_83A2683.jpg")
```

Adding Salt and Pepper Noise

```python
noise = np.random.randint(0, 2, img.shape, dtype=np.uint8)
noise[noise == 1] = 200
noisy_image = cv2.add(img, noise)
```

```python
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(122)
plt.imshow(noisy_image, cmap='gray')
plt.title('Noisy Image (Salt-and-Pepper)')
plt.show()
```

Applying Median Filter

```python
restored_image = cv2.medianBlur(noisy_image, 3)
```

```python
plt.imshow(restored_image, cmap='gray')
plt.title('Restored Image (Median Filter)')
plt.show()
```
