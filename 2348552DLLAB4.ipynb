```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('/content/image.jpeg')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
kernel = np.ones((3, 3), np.float32) / 9
```

```python
smoothed_image = cv2.filter2D(gray_image, -1, kernel)
```

```python
def apply_averaging_kernel(image, kernel):
    rows, cols = image.shape
    output = np.zeros((rows, cols), dtype=np.uint8)
    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
            output[i, j] = np.sum(image[i-1:i+2, j-1:j+2] * kernel)
    return output
```

```python
kernel = np.ones((3, 3), np.float32) / 9
```

```python
smoothed_image_ud = apply_averaging_kernel(gray_image, kernel)
```

```python
plt.figure(figsize=(20, 15))
plt.subplot(1, 3, 1), plt.imshow(gray_image, cmap='gray'), plt.title('Original')
plt.subplot(1, 3, 2), plt.imshow(smoothed_image, cmap='gray'), plt.title('Smoothed (OpenCV)')
plt.subplot(1, 3, 3), plt.imshow(smoothed_image_ud, cmap='gray'), plt.title('Smoothed (Custom)')
plt.show()
```
