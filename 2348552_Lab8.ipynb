```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread(r"C:\Users\sanja\Downloads\Gemini_Generated_Image(1).jpg",0)
```

1st order derivative method:Sobel Operator

```python
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
```

```python
sobel_combined = cv2.addWeighted(np.absolute(sobel_x), 0.5, np.absolute(sobel_y), 0.5, 0)
```

Second Order Derivated Method: Laplacian Filter

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
```

Optimum Edge Detection Method: Canny Edge Detector

```python
canny = cv2.Canny(img, 100, 200)
```

```python
plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(222)
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Operator')

plt.subplot(223)
plt.imshow(np.absolute(laplacian), cmap='gray')
plt.title('Laplacian Operator')

plt.subplot(224)
plt.imshow(canny, cmap='gray')
plt.title('Canny Edge Detector')

plt.show()
```
