```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/img.jpeg')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
equalized_image = cv2.equalizeHist(gray_image)
```

```python
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
```

```python
sharpening_kernel = np.array([[-1, -1, -1],
                              [-1, 9, -1],
                              [-1, -1, -1]])
sharpened_image = cv2.filter2D(image, -1, sharpening_kernel)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 3, 1), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(2, 3, 2), plt.imshow(gray_image, cmap='gray'), plt.title('Grayscale')
plt.subplot(2, 3, 3), plt.imshow(equalized_image, cmap='gray'), plt.title('Histogram Equalization')
plt.subplot(2, 3, 4), plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)), plt.title('Gaussian Blur')
plt.subplot(2, 3, 5), plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB)), plt.title('Sharpening')
plt.show()
```
